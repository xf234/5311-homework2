```python
# Select only the specified columns
selected_columns = ['Passengers', 'Fly_date']

# Make a copy of the DataFrame with only the selected columns
df_final = ord_df_o[selected_columns].copy()

# Extract the year from the 'Fly_date' column and add it as a new column 'Year'
df_final['Year'] = df_final['Fly_date'].str.split('-').str[0]

# Extract the month from the 'Fly_date' column and add it as a new column 'Month'
df_final['Month'] = df_final['Fly_date'].str.split('-').str[1]

# Drop the original 'Fly_date' column since we've extracted the year and month
df_final.drop(columns=['Fly_date'], inplace=True)


df_sum_passengers = df_final.groupby(['Year', 'Month']).agg({'Passengers': 'sum'}).reset_index()
# Display the first few rows of the selected data
df_sum_passengers
#df_sum_passengers.to_csv('/Users/xiaoyufan/Desktop/df_sum_passengers.csv', index=False)
```

```python
# Group by year and sum up the passengers
ord_passenger = df_final.groupby('Year')['Passengers'].sum().reset_index()

# Display the new DataFrame
ord_passenger
#ord_passenger.to_csv('/Users/xiaoyufan/Desktop/ord_passenger.csv', index=False)
```

```python
#who has most passengers
# Combine passengers for Origin_airport and Destination_airport
origin_passengers = df.groupby('Origin_airport')['Passengers'].sum().reset_index()
destination_passengers = df.groupby('Destination_airport')['Passengers'].sum().reset_index()

# Rename columns to be consistent
origin_passengers.rename(columns={'Origin_airport': 'Airport'}, inplace=True)
destination_passengers.rename(columns={'Destination_airport': 'Airport'}, inplace=True)

# Concatenate dataframes
total_passengers = pd.concat([origin_passengers, destination_passengers])

# Group by airport and sum the Passenger_count
total_passengers = total_passengers.groupby('Airport')['Passengers'].sum().reset_index()

# Find the airport with the most passengers
airport_with_most_passengers = total_passengers.loc[total_passengers['Passengers'].idxmax()]

print(total_passengers)
print("Airport with the most passengers:", airport_with_most_passengers['Airport'])
print("Number of passengers:", airport_with_most_passengers['Passengers'])
```

```python
#ATL: has most passengers
atl_psgr = df[(df['Destination_airport'] == 'ATL') | (df['Origin_airport'] == 'ATL')]
atl_psgr_delete_column = atl_psgr[selected_columns].copy()
atl_psgr


atl_psgr_delete_column['Year'] = atl_psgr_delete_column['Fly_date'].str.split('-').str[0]

atl_psgr_delete_column['Month'] = atl_psgr_delete_column['Fly_date'].str.split('-').str[1]


atl_psgr_delete_column.drop(columns=['Fly_date'], inplace=True)

atl_psgr_delete_column
sum_passengers_atl = atl_psgr_delete_column[atl_psgr_delete_column['Year'].astype(int) >= 2000]

sum_passengers_atl = sum_passengers_atl.groupby(['Year', 'Month'])['Passengers'].sum().reset_index()

print(sum_passengers_atl)
#sum_passengers.to_csv('/Users/xiaoyufan/Desktop/most_psgr_atl.csv', index=False)
```

```python
#find airport with most flights
# Combine flights for Origin_airport and Destination_airport
origin_flights = df.groupby('Origin_airport')['Flights'].sum().reset_index()
destination_flights = df.groupby('Destination_airport')['Flights'].sum().reset_index()

# Rename columns to be consistent
origin_flights.rename(columns={'Origin_airport': 'Airport'}, inplace=True)
destination_flights.rename(columns={'Destination_airport': 'Airport'}, inplace=True)

# Concatenate dataframes
total_flights = pd.concat([origin_flights, destination_flights])

# Group by airport and sum the Flights
total_flights = total_flights.groupby('Airport')['Flights'].sum().reset_index()

# Find the airport with the most flights
airport_with_most_flights = total_flights.loc[total_flights['Flights'].idxmax()]

print(total_flights)
print("Airport with the most flights:", airport_with_most_flights['Airport'])
print("Number of flights:", airport_with_most_flights['Flights'])
```

```python
#ORD has most flights, export flights of ord
new_selected_columns = ['Flights', 'Fly_date']
ord_psgr = df[(df['Destination_airport'] == 'ORD') | (df['Origin_airport'] == 'ORD')]
ord_psgr_delete_column = ord_psgr[new_selected_columns].copy()
ord_psgr


ord_psgr_delete_column['Year'] = ord_psgr_delete_column['Fly_date'].str.split('-').str[0]

ord_psgr_delete_column['Month'] = ord_psgr_delete_column['Fly_date'].str.split('-').str[1]


ord_psgr_delete_column.drop(columns=['Fly_date'], inplace=True)
ord_psgr_delete_column

sum_flights_ord = ord_psgr_delete_column.groupby(['Year', 'Month'])['Flights'].sum().reset_index()
sum_flights_ord = sum_flights_ord[sum_flights_ord['Year'].astype(int) >= 2000]
print(sum_flights_ord)
sum_flights_ord.to_csv('/Users/xiaoyufan/Desktop/most_flights_ord.csv', index=False)
```

```python
#find sum flights of each month
df_selected = df[new_selected_columns].copy()

# Step 2: Parse 'Fly_date' and add 'Year' and 'Month' columns
df_selected['Year'] = pd.to_datetime(df_selected['Fly_date']).dt.year
df_selected['Month'] = pd.to_datetime(df_selected['Fly_date']).dt.month

# Step 3: Delete the 'Fly_date' column
df_selected.drop(columns=['Fly_date'], inplace=True)

# Step 4: Sum up flights where they have the same month and year
sum_flights = df_selected.groupby(['Year', 'Month']).sum().reset_index()
sum_flights = sum_flights[sum_flights['Year'].astype(int) >= 2000]
print(sum_flights)
#sum_flights.to_csv('/Users/xiaoyufan/Desktop/sum_flights.csv', index=False)
```
