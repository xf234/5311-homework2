## Pop-up Box Info

```python

# Assuming `df` is your original DataFrame

# Combine origin and destination airport codes to define a unique route
df['route'] = df['Origin_airport'] + '-' + df['Destination_airport']

# Calculate the number of unique routes and passengers per airport per year
routes_per_airport = df.groupby(['year', 'Origin_airport', 'Origin_city', 'Origin_population']).agg(
    number_of_routes=('route', 'nunique'),
    total_passengers=('Passengers', 'sum')
).reset_index()

# Find the most popular route per airport per year
most_popular_route = df.groupby(['year', 'Origin_airport'])['route'].apply(
    lambda x: x.value_counts().index[0]
).reset_index(name='most_popular_route')

# Find the number of passengers for the most popular route
passengers_for_popular_route = df.groupby(['year', 'Origin_airport', 'route']).agg(
    passengers=('Passengers', 'sum')
).reset_index()

# Merge to get the most popular route's passenger count
popular_route_passengers = pd.merge(
    most_popular_route,
    passengers_for_popular_route,
    left_on=['year', 'Origin_airport', 'most_popular_route'],
    right_on=['year', 'Origin_airport', 'route'],
    how='left'
)

# Merge with the routes_per_airport to consolidate data
consolidated = pd.merge(
    routes_per_airport,
    popular_route_passengers[['year', 'Origin_airport', 'most_popular_route', 'passengers']],
    on=['year', 'Origin_airport'],
    how='left'
)

# Sort and rank airports within each year by the number of routes
consolidated['rank'] = consolidated.groupby('year')['number_of_routes'].rank("dense", ascending=False)

# Filter to get the top 15 airports for each year
top_airports = consolidated[consolidated['rank'] <= 15]

# Rename columns for clarity
top_airports.rename(columns={
    'Origin_airport': 'airport',
    'Origin_city': 'city',
    'Origin_population': 'population',
    'most_popular_route': 'most_popular_route',
    'passengers': 'number_of_passengers'
}, inplace=True)

# Select only the required columns
final_df = top_airports[['year', 'airport', 'city', 'population', 'most_popular_route', 'number_of_passengers']]

print(final_df)
```

```python
final_df.to_csv("top_15_airport.csv")
```

## Airport Traffic

```python
origin_df = df_new[['year', 'Origin_airport', 'Passengers', 'Org_airport_lat', 'Org_airport_long']].rename(columns={'Origin_airport': 'Airport', 'Org_airport_lat': 'Lat', 'Org_airport_long': 'Long'})
destination_df = df_new[['year', 'Destination_airport', 'Passengers', 'Dest_airport_lat', 'Dest_airport_long']].rename(columns={'Destination_airport': 'Airport', 'Dest_airport_lat': 'Lat', 'Dest_airport_long': 'Long'})

origin_traffic = origin_df.groupby(['year', 'Airport']).agg({'Passengers': 'sum', 'Lat': 'first', 'Long': 'first'}).reset_index()
destination_traffic = destination_df.groupby(['year', 'Airport']).agg({'Passengers': 'sum', 'Lat': 'first', 'Long': 'first'}).reset_index()

combined_df = pd.concat([origin_traffic, destination_traffic])

total_traffic = combined_df.groupby(['year', 'Airport']).agg({'Passengers': 'sum', 'Lat': 'mean', 'Long': 'mean'}).reset_index()

total_traffic = total_traffic.rename(columns={'Passengers': 'Traffic', 'Lat': 'Airport_Lat', 'Long': 'Airport_Long'})

total_traffic = total_traffic.dropna()
total_traffic = total_traffic[total_traffic["Traffic"] != 0]
total_traffic = total_traffic.reset_index()

print(total_traffic)
```
